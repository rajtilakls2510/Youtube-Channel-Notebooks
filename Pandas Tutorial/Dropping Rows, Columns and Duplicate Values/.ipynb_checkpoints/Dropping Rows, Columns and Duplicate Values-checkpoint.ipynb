```python
import pandas as pd
```

### Let's see the data first

```python
matches=pd.read_csv('matches.csv')
```

```python
matches.head()
```

## drop()

```python
matches.drop(labels=[1,4,6],axis='index').head()
```

```python
# Version>=0.21
matches.drop(index=[1,4,6]).head(10)
```

## Dropping columns

```python
matches.drop(labels=['city','team2'],axis='columns').head()
```

#### Dropping rows and columns simultaneously 

```python
matches.drop(index=[2,3,4],columns=['season','date']).head()
```

## Drop Duplicates

```python
matches.shape
```

```python
matches.drop_duplicates().shape
```

```python
kkr_csk=matches[(matches["winner"]=="Kolkata Knight Riders") | (matches["winner"]=="Chennai Super Kings")]
```

```python
kkr_csk.sample(10)
```

```python
kkr_csk.drop_duplicates(subset=['winner'],keep='last')
```
