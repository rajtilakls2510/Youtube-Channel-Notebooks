### Extracting a single column

### Extracting multiple columns

```python
weather[['Date','Location','Rainfall']]
```

### Select all rows where Location is Sydney

```python
weather[weather['Location']=='Sydney']
```

```python
weather[mask1 & mask2]
```

```python
weather['AvgTemp']=(weather['MinTemp']+weather['MaxTemp'])/2
```
