## Merge

```python
df1 = pd.DataFrame({'key': [1, 2, 3, 5],
                   'value': [213,456,789,658]})
df2 = pd.DataFrame({'key': [5, 6, 7, 8],
                    'value': [756,845,956,356]})
df1
```

```python
df2
```

```python
df1.merge(df2,on='key')
```

```python
df1 = pd.DataFrame({'lkey': [1, 2, 3, 5],
                   'value': [213,456,789,658]})
df2 = pd.DataFrame({'rkey': [5, 6, 7, 8],
                    'value': [756,845,956,356]})
df1
```

```python
df2
```

```python
df1.merge(df2,left_on='lkey',right_on='rkey')
```

### Let's see the 'how' parameter

```python
df1.merge(df2,left_on='lkey',right_on='rkey',how='inner')
```

```python
df1.merge(df2,left_on='lkey',right_on='rkey',how='left')
```

```python
df1.merge(df2,left_on='lkey',right_on='rkey',how='right')
```

```python
df1.merge(df2,left_on='lkey',right_on='rkey',how='outer')
```

## Loading IPL Data

```python
matches=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
matches.head()
```

```python
delivery.head()
```

## Orange Cap Winners of all the season

```python
merged=matches.merge(delivery,left_on='id',right_on='match_id',how='inner')
```

```python
merged.groupby(['season','batsman'],as_index=False)['batsman_runs'].agg('sum').sort_values('batsman_runs',ascending=False).drop_duplicates('season',keep='first').sort_values('season')
```
