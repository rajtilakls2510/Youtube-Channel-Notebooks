# How do we read data in Pandas?

```python
import pandas as pd
```

### Reading Local Dataset

```python
pd.read_csv('titanic.csv')
```

```python
pd.read_csv('data/food.csv')
```

### Reading Data from Url

```python
pd.read_csv('http://bit.ly/uforeports')
```

### Reading data without Headers

```python
columns=['Name','Gender','City','Frequency','Item','Spends']
```

```python
pd.read_csv('food1.csv',header=None,names=columns)
```

### Reading Data with different separators

```python
columns=['user_id','age','gender','occupation','zip_codes']
```

```python
pd.read_csv('http://bit.ly/movieusers',sep='|',header=None,names=columns)
```
