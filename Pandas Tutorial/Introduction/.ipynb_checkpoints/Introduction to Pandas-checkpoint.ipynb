# Pandas

In computer programming, [pandas](https://pandas.pydata.org/docs/#module-pandas) is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series.

Source: [Wikipedia](https://en.wikipedia.org/wiki/Pandas_(software))

## The 2 Main Data Structures it provides:

### 1. DataFrame

Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.

Source: [GeeksForGeeks](https://www.geeksforgeeks.org/python-pandas-dataframe/)


### 2. Series

Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively called index. Pandas Series is nothing but a column in an excel sheet.

Source: [GeeksForGeeks](https://www.geeksforgeeks.org/python-pandas-series/)
