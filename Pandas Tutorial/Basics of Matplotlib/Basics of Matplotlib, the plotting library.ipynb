```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Line Charts

```python
x=[25,10,163,4,79]
y=[1,2,3,4,5]
```

```python
plt.plot(x,y)
```

```python
year=range(2012,2020)
salary=[200,100,500,600,700,800,900,1000]
```

```python
plt.plot(year,salary)
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('Year vs Salary of ABC Company')
plt.show()
```

## Scatter Plots

```python
x=np.random.randint(10,1000,20)
y=np.random.randint(1,10,20)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.show()
```

```python
year=range(2012,2020)
salary1=[200,100,500,600,700,800,900,1000]
salary2=[500,700,800,900,1000,1200,1300,1400]
```

```python
plt.scatter(year,salary1,label='Company 1')
plt.scatter(year,salary2,label='Company 2')
plt.legend()
```

## Pie Charts

```python
expenditure=[10,20,5,4,30]
dept=['Research','Technology','Production','Analysis','Delivery']
```

```python
plt.pie(expenditure,labels=dept,autopct='%.1f%%')
plt.show()
```

## Bar Charts

```python
expenditure=[10,20,5,4,30]
dept=['Research','Technology','Production','Analysis','Delivery']
```

```python
plt.bar(dept,expenditure)
plt.xlabel('Departments')
plt.ylabel('Expenditure')
plt.title('Company Departmental Expenditures')
plt.show()
```
