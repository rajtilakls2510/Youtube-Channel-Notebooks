```python
import numpy as np
```

# Generating Random Arrays

### random.random()

```python
np.random.random(size=(15,4)) 
```

### random.randint()

```python
np.random.randint(low=10,high=101,size=(19,23))
```

### random.uniform()

```python
np.random.uniform(low=10.56, high=12.40,size=(10,2))
```

### random.choice()

```python
a=np.arange(50)
a
```

```python
np.random.choice(a,size=(10,20))
```

# Shuffling Arrays

### random.shuffle()

```python
a
```

```python
np.random.shuffle(a)
a
```

```python
np.random.permutation(a)
```

# Generating Standard Normal Distribution

```python
np.random.standard_normal(size=100000).mean()
```

```python
np.random.standard_normal(size=100000).std()
```
